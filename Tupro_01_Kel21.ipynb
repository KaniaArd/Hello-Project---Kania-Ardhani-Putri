```python
import random
import math

def generateKromosom(panjangKromosom): #membuat kromosom berdasarkan panjangKromosom (mengambil 0 dan 1 secara acak)
    kromosom = [] 

    for i in range(panjangKromosom): 
        kromosom.append(random.randint(0,1)) 
    return kromosom


def generatePopulasi(ukuranPopulasi, panjangKromosom): #membuat populasi berdasarkan ukuranPopulasi
    pop = [] 

    for i in range(ukuranPopulasi): 
        pop.append(generateKromosom(panjangKromosom)) 
    return pop
     

def dekodeKromosom(kromosom, panjangKromosom): # mengkonversi nilai kromosom menjadi nilai x dan y
    xMin, xMax = (-5, 5)
    yMin, yMax = (-5, 5)
    x = 0
    y = 0
    z = 0
    n = (panjangKromosom) // 2

    for i in range(0, n):
        z += 2**-(i+1)
    for i in range(0, n):
        x += kromosom[i] * 2 ** -(i+1)
        y += kromosom[n + i] * 2 ** -(i+1)
    x = xMin + ((xMax - xMin) / z) * x
    y = yMin + ((yMax - yMin) / z) * y
    return [x, y]
   

def fungsiObjektif(x, y): #menghitung nilai objektif, yakni hasil perhitungan
    return ((math.cos(x)+math.sin(y))**2) / (x**2 + y**2)


def fungsi(x, y): #menghitung nilai minimisasi dari fungsi
    return 1 / (0.01 + ((math.cos(x)+math.sin(y))**2) / (x**2 + y**2))
  

def fitness(populasi, panjangKromosom): #memasukkan semua nilai fitness ke dalam array
    hasil = []

    for i in populasi:
        hasil.append(fungsi(*dekodeKromosom(i, panjangKromosom)))
    return hasil
    

def tournament(populasi, ukuranPopulasi, ukuranTour, panjangKromosom): #seleksi orang tua menggunakan metode tournament
    hasil = []

    for _ in range(ukuranTour):
        temp = populasi[random.randint(0, ukuranPopulasi - 1)]
        if hasil == [] or fungsi(dekodeKromosom(temp, panjangKromosom)[0], 
        dekodeKromosom(temp, panjangKromosom)[0]) < fungsi(dekodeKromosom(hasil, panjangKromosom)[0], 
        dekodeKromosom(hasil, panjangKromosom)[0]):
            hasil = temp
    return hasil

    
def crossover(parentA, parentB, pc, panjangKromosom): #melakukan crossover dengan metode two point crossover
    value = random.uniform(0, 1)
 
    if value < pc:
        mv = random.randint(0, panjangKromosom-1)
        for i in range(mv):
            parentA[i], parentB[i] = parentB[i], parentA[i]
    return [parentA, parentB]

    
def mutasi(keturunan, pm, panjangKromosom): #melakukan mutasi keturunan
    value = random.uniform(0, 1)

    if value < pm:
        keturunan[0][random.randint(0, panjangKromosom - 1)] = random.randint(0,1) 
        keturunan[1][random.randint(0, panjangKromosom - 1)] = random.randint(0,1) 
    return keturunan


def elitisme(fit): #mencari 2 nilai terbaik pada fitness
    elit1, elit2 = (0,0)

    for i in range(len(fit)):
        if fit[i] < fit[elit1]:
            elit2 = elit1
            elit1 = i
    return [elit1, elit2]


def generasialReplacement(): #melakukan seleksi survivor dengan metode generational replacement
    panjangKromosom = 8
    ukuranPopulasi = 100
    pc = 0.8 #Crossover Probability
    pm = 0.05 #Mutation Probability
    generasi = 100
    ukuranTour = 5 
    
    populasi = generatePopulasi(ukuranPopulasi, panjangKromosom)
    for _ in range(generasi):
        fit = fitness(populasi, panjangKromosom)
        populasiBaru = []
        elite1, elite2 = elitisme(fit)
        populasiBaru.append(populasi[elite1])
        populasiBaru.append(populasi[elite2])
        for _ in range(0, ukuranPopulasi - 2, 2):
            parentA = tournament(populasi, ukuranPopulasi, ukuranTour, panjangKromosom)
            parentB = tournament(populasi, ukuranPopulasi, ukuranTour, panjangKromosom)
            while(parentA == parentB):
                parentB = tournament(populasi, ukuranPopulasi, ukuranTour, panjangKromosom)  
            keturunan = crossover(parentA[:], parentB[:], pc, panjangKromosom)
            keturunan = mutasi(keturunan, pm, panjangKromosom)
            populasiBaru.extend(keturunan)
    populasi = populasiBaru
    printhasil(populasi, panjangKromosom, ukuranPopulasi, generasi, fit)


def printhasil(populasi, panjangKromosom, ukuranPopulasi, generasi, fit):
    idx = fit.index(max(fit))
    dekode = dekodeKromosom(populasi[idx], panjangKromosom)

    print("Panjang kromosom\t: ", panjangKromosom)
    print("Ukuran populasi\t\t: ", ukuranPopulasi)
    print("Generasi\t\t: ", generasi)
    print("===============Hasil Genetic Algorithm===============")
    print("Kromosom terbaik\t: ", populasi[idx])
    print("Nilai fitness\t\t: ", fit[idx])
    print("Nilai fungsi\t\t: ", fungsiObjektif(dekode[0], dekode[1]))
    print("Nilai X\t\t\t: ", dekode[0])
    print("Nilai y\t\t\t: ", dekode[1])

generasialReplacement()
```
